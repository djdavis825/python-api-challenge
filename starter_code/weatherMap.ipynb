```python
import numpy as np
```

```python
# set range of lat/long
latRange = (24, 50)
lngRange = (-125, -66.9)
```

```python
# generate a series of values in a range
lats = np.random.uniform(latRange[0], latRange[1], size=50)
long = np.random.uniform(lngRange[0], lngRange[1], size=50)

latLongs = zip(lats,long)
```

```python
cityNames = []

for coords in latLongs:
    #get a city name
    cityName = citipy.nearest_city(coords[0], coords[1]).city_name
    
    if cityName not in cityNames:
        cityNames.append(cityName)

for c in cityNames:
    print(c)
```

```python
# Make dataframe
temperatures = []
lats = []
long = []
cities = []

queryURL = f"{url}appid={weather_api_key}&units={units}&q="

for city in cityNames:
    
    try:
        # new query string
        cityQuery = f"{queryURL}{city}"
        
        #jsonify the response
        response = requests.get(cityQuery).json()
        
        if response['coord']['lat'] > 50 or response['coord']['lat'] < 24:
            pass
        elif response['coord']['lon'] < -125 or response['coord']['lon'] > -66.9:
            pass
        elif response['sys']['country'] != 'US':
            pass
        else:
            #add to lists
            temperatures.append(response['main']['temp'])
            lats.append(response['coord']['lat'])
            long.append(response['coord']['lon'])
            cities.append(response['name'])
        
    except:
        pass
#build dataframe
cityDataDF = pd.DataFrame(
    {
        "City Name": cities,
        "Fahernheit Temp": temperatures,
        "Latitude": lats,
        "Longitude": long
    }
)

cityDataDF
```
